## Hyperspheres

Hyperspheres are generalisations of circles in an arbitrary dimension $n$ with the special cases of circles ($n=2$) and spheres ($n=3$). The definition of a hypersphere of radius $s>0$ centred at the origin is the set of points $x$ with distance $s$ from the origin or in other words all vectors $x$ of length or Euclidean norm $\|x\|=s$:

$$
S = \left\{\mathbf x \in \mathbb{R}^n \, : \, \|\mathbf x\| = \sqrt{x^2_1 + \dots + x_n^2} = s \right\}
$$

Alternatively, we can think of the same circle as a *unit circle*, i.e., of radius $1$, with respect to a scaled coordinate system, where coordinates represent multiples of $s$:

$$
\begin{aligned}
S &= \left\{\mathbf x \in \mathbb{R}^n \, : \, \|\mathbf x\|^2 = x^2_1 + \dots + x_n^2 = s^2 \right\}\\
&=\left\{\mathbf x \in \mathbb{R}^n \, : \, \|s^{-1}\mathbf x\|^2 = \left(\frac{x_1}{s}\right)^2 + \dots + \left(\frac{x_n}{s}\right)^2 = 1 \right\}
\end{aligned}
$$

```python
import numpy as np
from matplotlib.patches import Circle
import matplotlib.pyplot as plt
# from plotting import centre_spines, superimpose_axis
import plotting

fig, ax = plt.subplots(figsize=(6, 6))
s = 2.5
circle = Circle(xy=(0, 0), radius=s, edgecolor='black', linewidth=1.2, fill=False)
ax.add_patch(circle)
phi = np.pi/3
x = np.array([s*np.cos(phi), s*np.sin(phi)])
ax.plot([0, x[0]], [0, x[1]], linestyle='--', c='k')
ax.scatter(s*np.cos(phi), s*np.sin(phi), c='red')
ax.centre_spines()
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
for axis, side in (['x', 'y'], ['positive', 'negative']):
    ax.superimpose_axis([-1.5*s, -1.0*s, -.5*s, .5*s, 1.0*s, 1.5*s], [-1.5, -1, -.5, .5, 1, 1.5], axis=axis, color='blue', fontsize=9)
plt.show()
```

## Standard Form Ellipsoids

The last form is easily generalised to **ellipsoids**. These are generalisations of ellipses from two dimensions ($n=2$) to arbitrary dimension n. They arise from replacing the uniform scaling factor $s$ in the circle definition with different factors $s_1, \dots, s_n$ per coordinate axis:

$$E = \left\{\mathbf x \in \mathbb{R}^n \,:\, \left(\frac{x_1}{s_1}\right)^2 + \dots + \left(\frac{x_n}{l_n}\right)^2 = 1\right\}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

a1, a2 = 2.5, 1.5
fig, ax = plt.subplots(figsize=(6, 6))
e = Ellipse((0, 0), 2*a1, 2*a2, fill=False, linewidth=1.2)
ax.add_patch(e)
phi = np.pi/3
x = np.array([a1*np.cos(phi), a2*np.sin(phi)])
ax.scatter(x[0], x[1], c='red')
ax.plot([0, x[0]], [0, x[1]], linestyle='--', c='k')
ax.annotate(r'$\mathbf{x}$', x, textcoords='offset points', xytext=( 4, 8))
ax.centre_spines()
ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-2.5, 2.5)
for axis, s in zip(['x', 'y'], [a1, a2]):
    ax.superimpose_axis([-1.5*s, -1.0*s, -.5*s, .5*s, 1.0*s, 1.5*s], [-1.5, -1, -.5, .5, 1, 1.5], color='blue', axis=axis)
    # ax.add_artist(SuperimposedAxis([-1.5*s, -1.0*s, -.5*s, .5*s, 1.0*s, 1.5*s], [-1.5, -1, -.5, .5, 1, 1.5], color='blue', axis=axis))
    # superimpose_axis(ax, [-1.5*s, -1.0*s, -.5*s, .5*s, 1.0*s, 1.5*s], [-1.5, -1, -.5, .5, 1, 1.5], axis=axis, color='blue', fontsize=9)

plt.tight_layout()
plt.show()
```
